```python
dwell_Q1_24 = pd.read_csv(r"C:\Users\qxnq723\Desktop\Project 1\HUQ Data\FW_ Dwell Time data to analyse please\Q1_24_Bradford_Dwell.csv")
dwell_Q1_25 = pd.read_csv(r"C:\Users\qxnq723\Desktop\Project 1\HUQ Data\FW_ Dwell Time data to analyse please\Q1_25_Bradford_Dwell.csv")
dwell_Q2_24 = pd.read_csv(r"C:\Users\qxnq723\Desktop\Project 1\HUQ Data\FW_ Dwell Time data to analyse please\Q2_24_Bradford_Dwell.csv")
dwell_Q2_25 = pd.read_csv(r"C:\Users\qxnq723\Desktop\Project 1\HUQ Data\FW_ Dwell Time data to analyse please\Q2_25_Bradford_Dwell.csv")
dwell_Q3_24 = pd.read_csv(r"C:\Users\qxnq723\Desktop\Project 1\HUQ Data\FW_ Dwell Time data to analyse please\Q3_24_Bradford_Dwell.csv")
dwell_Q3_25 = pd.read_csv(r"C:\Users\qxnq723\Desktop\Project 1\HUQ Data\FW_ Dwell Time data to analyse please\Q3_25_Bradford_Dwell.csv")
dwell_Q4_24 = pd.read_csv(r"C:\Users\qxnq723\Desktop\Project 1\HUQ Data\FW_ Dwell Time data to analyse please\Q4_24_Bradford_Dwell.csv")
dwell_Q4_25 = pd.read_csv(r"C:\Users\qxnq723\Desktop\Project 1\HUQ Data\FW_ Dwell Time data to analyse please\Q4_25_Bradford_Dwell.csv")
```

```python
#Add year and quarter columns to each dataset
dwell_Q1_24['year'] = 2024
dwell_Q1_24['quarter'] = 1

dwell_Q1_25['year'] = 2025
dwell_Q1_25['quarter'] = 1

dwell_Q2_24['year'] = 2024
dwell_Q2_24['quarter'] = 2

dwell_Q2_25['year'] = 2025
dwell_Q2_25['quarter'] = 2

dwell_Q3_24['year'] = 2024
dwell_Q3_24['quarter'] = 3

dwell_Q3_25['year'] = 2025
dwell_Q3_25['quarter'] = 3

dwell_Q4_24['year'] = 2024
dwell_Q4_24['quarter'] = 4

#That quarter has different format so standardize it
dwell_Q4_25['year'] = 2025
dwell_Q4_25['quarter'] = 4
dwell_Q4_25 = dwell_Q4_25.drop(columns= ['Value (Weekdays only)', 'Value (Weekends only)'])
dwell_Q4_25 = dwell_Q4_25.rename(columns={'Value (All days of week)': 'Value'})
dwell_Q4_25 = dwell_Q4_25.rename(columns={'area': 'Area'})
dwell_Q4_25 = dwell_Q4_25.rename(columns={'common.dataType': 'Data Type'})
```

```python
#Concatenate the datasets together
dwell = pd.concat([dwell_Q1_24,
                   dwell_Q1_25,
                   dwell_Q2_24,
                   dwell_Q2_25,
                   dwell_Q3_24,
                   dwell_Q3_25,
                   dwell_Q4_24,
                   dwell_Q4_25
                    ], axis=0)

dwell.info()
```

```python
dwell['Data Type'].unique()
```

```python
#Separate the 'Dwell Time Distribution' from the 'Dwell Time by Day of Week'

dwell_distribution = dwell[dwell['Data Type'] == 'Dwell Time Distribution']

dwell_week = dwell[dwell['Data Type'] == 'Dwell Time by Day of Week']
```

```python
dwell_week.head(20)
```

```python
dwell_week['Area'].unique()
```

```python
#Take out the minutes string
dwell_week['Value'] = dwell_week['Value'].str.replace(' minutes', '')
#Convert Value column to numerical instead of object
dwell_week['Value'] = pd.to_numeric(dwell_week['Value'], errors='coerce')
dwell_week.head()
```

```python
# Get averages of days in each quarter
quarter_summary = (
    dwell_week
    .groupby(['quarter', 'year'])['Value']
    .mean()
    .unstack('year'))

#Calculate percentage chnage in quarters between 2024 and 2025
quarter_summary['pct_change_24-25'] = ((quarter_summary[2025] - quarter_summary[2024]) / quarter_summary[2024] * 100)

quarter_summary
```

```python
#Get average of quarters over years
data_per_year = (dwell_week.groupby(['year'])['Value'].mean())
print(data_per_year)
```
